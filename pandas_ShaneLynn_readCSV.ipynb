# Deleting rows and columns (drop)

```python
data = data.drop("Area", axis=1)
```

```python
data = data.drop(columns="Area Code")
```

```python
data.drop("Item Code", axis=1, inplace=True)
```

```python
data = data.drop(["Y2001", "Y2002", "Y2003"], axis=1)
```

```python
data.head(3)
```

```python
data.drop([0,1], axis=0).head(3) #delete rows with index of 0 and 1
```

```python
data = data.set_index("Element")
data.drop("Food", axis=0)
```

```python
data = data.reset_index()
```

```python
data.iloc[5:,]
```

# Renaming columns

```python
data.rename(columns={"Element": "Elements"})
```

```python
data.rename(columns={"Unit":"Units", "Item":"Items"}, inplace=True)
```

```python
data.head()
```

```python
data.rename(columns=str.lower)
```

```python
data.rename(columns=lambda x: x.upper().replace(' ', '_')) # Quickly lowercase and camelcase all column names in a DataFrame
```
